```python
# E-Commerce Transaction Data - Exploratory Data Analysis (EDA)

# Goal:
# - Clean the dataset
# - Handle missing values
# - Analyze customer behavior and product performance
# - Generate visual insights using Python libraries: pandas, matplotlib, seaborn

# --- 1. Import libraries ---
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

pd.set_option('display.max_columns', 50)
sns.set(style="whitegrid")

# --- 2. Load Data ---
# Using your uploaded file 'ecommercedata.csv' in the current directory
df = pd.read_csv("ecommercedata.csv", encoding='cp1252')


print("Initial data preview:")
display(df.head())

# --- 3. Data Overview ---
print(f"Dataset shape: {df.shape}")
print("\nData info:")
print(df.info())

print("\nSummary statistics:")
display(df.describe())

# Check duplicates
num_duplicates = df.duplicated().sum()
print(f"\nNumber of duplicate rows: {num_duplicates}")
df = df.drop_duplicates()

# Check missing values
missing_values = df.isnull().sum()
print("\nMissing values by column:")
print(missing_values)

# Visualize missing values heatmap
plt.figure(figsize=(8,4))
sns.heatmap(df.isnull(), cbar=False, yticklabels=False)
plt.title('Missing Values Heatmap')
plt.show()

# --- 4. Data Cleaning ---
# Drop rows where CustomerID or Description is missing because these are essential columns
df_clean = df.dropna(subset=['CustomerID', 'Description'])

print(f"\nData shape after dropping rows with missing CustomerID or Description: {df_clean.shape}")

# Convert InvoiceDate to datetime
df_clean['InvoiceDate'] = pd.to_datetime(df_clean['InvoiceDate'], errors='coerce')
# Drop rows with invalid InvoiceDate
df_clean = df_clean.dropna(subset=['InvoiceDate'])

# Create TotalPrice column = Quantity * UnitPrice
df_clean['TotalPrice'] = df_clean['Quantity'] * df_clean['UnitPrice']

# --- 5. Feature Engineering ---
df_clean['InvoiceMonth'] = df_clean['InvoiceDate'].dt.to_period('M')

# --- 6. Customer Behavior Analysis ---

# a) Number of unique orders per customer
orders_per_customer = df_clean.groupby('CustomerID')['InvoiceNo'].nunique()

plt.figure(figsize=(8,4))
sns.histplot(orders_per_customer, bins=30, kde=False)
plt.title('Number of Orders per Customer')
plt.xlabel('Number of Orders')
plt.ylabel('Count of Customers')
plt.show()

# b) Distribution of total spending per customer
spending_per_customer = df_clean.groupby('CustomerID')['TotalPrice'].sum()

plt.figure(figsize=(8,4))
sns.histplot(spending_per_customer, bins=30, kde=False)
plt.title('Total Spending per Customer')
plt.xlabel('Total Spending')
plt.ylabel('Count of Customers')
plt.show()

# c) Top 10 Countries by number of orders
orders_by_country = df_clean.groupby('Country')['InvoiceNo'].nunique().sort_values(ascending=False).head(10)

plt.figure(figsize=(10,5))
sns.barplot(x=orders_by_country.values, y=orders_by_country.index, palette='viridis')
plt.title('Top 10 Countries by Number of Orders')
plt.xlabel('Number of Orders')
plt.ylabel('Country')
plt.show()

# --- 7. RFM (Recency-Frequency-Monetary) Analysis ---

# Define snapshot date as one day after last invoice date
snapshot_date = df_clean['InvoiceDate'].max() + pd.Timedelta(days=1)

rfm = df_clean.groupby('CustomerID').agg({
    'InvoiceDate': lambda x: (snapshot_date - x.max()).days,
    'InvoiceNo': 'nunique',
    'TotalPrice': 'sum'
}).rename(columns={'InvoiceDate': 'Recency',
                   'InvoiceNo': 'Frequency',
                   'TotalPrice': 'MonetaryValue'})

print("\nRFM table sample:")
display(rfm.head())

# Plot histograms of RFM metrics
plt.figure(figsize=(15,4))
plt.subplot(1,3,1)
sns.histplot(rfm['Recency'], bins=30, kde=False)
plt.title('Recency Distribution')
plt.xlabel('Days since last purchase')

plt.subplot(1,3,2)
sns.histplot(rfm['Frequency'], bins=30, kde=False)
plt.title('Frequency Distribution')
plt.xlabel('Number of Orders')

plt.subplot(1,3,3)
sns.histplot(rfm['MonetaryValue'], bins=30, kde=False)
plt.title('Monetary Value Distribution')
plt.xlabel('Total Spending')

plt.tight_layout()
plt.show()

# --- 8. Product Performance Analysis ---

# Top 10 products by total revenue
top_products = df_clean.groupby('Description')['TotalPrice'].sum().sort_values(ascending=False).head(10)

plt.figure(figsize=(12,6))
sns.barplot(x=top_products.values, y=top_products.index, palette='magma')
plt.title('Top 10 Products by Revenue')
plt.xlabel('Revenue')
plt.ylabel('Product')
plt.show()

# --- Returns and Negative Quantities ---
returns = df_clean[df_clean['Quantity'] < 0]
return_rate = len(returns) / len(df_clean) * 100
print(f"Return rate: {return_rate:.2f}%")

top_returned_products = returns['Description'].value_counts().head(10)

plt.figure(figsize=(10,5))
sns.barplot(x=top_returned_products.values, y=top_returned_products.index, palette='rocket')
plt.title('Top 10 Returned Products')
plt.xlabel('Number of Returns')
plt.ylabel('Product')
plt.show()

# --- 9. Monthly Sales Trend ---

monthly_sales = df_clean.groupby('InvoiceMonth')['TotalPrice'].sum()

plt.figure(figsize=(12,5))
monthly_sales.plot(marker='o')
plt.title('Monthly Total Sales Revenue')
plt.ylabel('Revenue')
plt.xlabel('Month')
plt.grid(True)
plt.show()

# --- 10. Optional: Save cleaned data ---
df_clean.to_csv("cleaned_ecommercedata.csv", index=False)

print("EDA complete. Cleaned data saved as 'cleaned_ecommercedata.csv'.")
# --- Data Cleaning and Handling Missing Values ---

# Drop rows with missing CustomerID or Description (these are essential for the analysis)
df_clean = df.dropna(subset=['CustomerID', 'Description'])

# Convert 'InvoiceDate' to datetime; invalid entries become NaT and are dropped
df_clean['InvoiceDate'] = pd.to_datetime(df_clean['InvoiceDate'], errors='coerce')
df_clean = df_clean.dropna(subset=['InvoiceDate'])

# Create a 'TotalPrice' column (Quantity * UnitPrice)
df_clean['TotalPrice'] = df_clean['Quantity'] * df_clean['UnitPrice']

# Show shape & check for remaining missing values
print("Shape after cleaning:", df_clean.shape)
print("Missing values after cleaning:\n", df_clean.isnull().sum())

# Preview the cleaned data
df_clean.head()
# --- Exploratory Visualizations: Customer Behavior and Product Performance ---

import matplotlib.ticker as mtick

# 1. Customer Behavior: Number of Orders per Customer
orders_per_customer = df_clean.groupby('CustomerID')['InvoiceNo'].nunique()

plt.figure(figsize=(8,4))
sns.histplot(orders_per_customer, bins=30, color='skyblue', kde=False)
plt.title('Distribution of Number of Orders per Customer')
plt.xlabel('Number of Orders')
plt.ylabel('Number of Customers')
plt.show()

# 2. Customer Behavior: Total Spending per Customer
spending_per_customer = df_clean.groupby('CustomerID')['TotalPrice'].sum()

plt.figure(figsize=(8,4))
sns.histplot(spending_per_customer, bins=30, color='lightgreen', kde=False)
plt.title('Distribution of Total Spending per Customer')
plt.xlabel('Total Spending ()')
plt.ylabel('Number of Customers')
plt.show()

# 3. Product Performance: Top 10 Best-Selling Products by Revenue
top_products = df_clean.groupby('Description')['TotalPrice'].sum().sort_values(ascending=False).head(10)

plt.figure(figsize=(12,6))
sns.barplot(x=top_products.values, y=top_products.index, palette='viridis')
plt.title('Top 10 Products by Total Revenue')
plt.xlabel('Revenue ()')
plt.ylabel('Product Description')
plt.show()

# 4. Product Performance: Return Analysis (Negative Quantity Transactions)
returns = df_clean[df_clean['Quantity'] < 0]
return_rate = len(returns) / len(df_clean) * 100

print(f"Return rate: {return_rate:.2f}% of transactions")

top_returned_products = returns['Description'].value_counts().head(10)

plt.figure(figsize=(10,5))
sns.barplot(x=top_returned_products.values, y=top_returned_products.index, palette='rocket')
plt.title('Top 10 Returned Products by Number of Return Transactions')
plt.xlabel('Number of Returns')
plt.ylabel('Product Description')
plt.show()

# --- Time Series and Geographic Analysis ---

# 1. Monthly Sales Trend (Total Revenue by Month)
df_clean['InvoiceMonth'] = df_clean['InvoiceDate'].dt.to_period('M')
monthly_sales = df_clean.groupby('InvoiceMonth')['TotalPrice'].sum()

plt.figure(figsize=(12,5))
monthly_sales.plot(marker='o', linestyle='-')
plt.title('Monthly Sales Revenue Trend')
plt.xlabel('Month')
plt.ylabel('Revenue ()')
plt.grid(True)
plt.show()

# 2. Sales Count by Country (Top 10 Countries)
orders_by_country = df_clean.groupby('Country')['InvoiceNo'].nunique().sort_values(ascending=False).head(10)

plt.figure(figsize=(10,5))
sns.barplot(x=orders_by_country.values, y=orders_by_country.index, palette='coolwarm')
plt.title('Top 10 Countries by Number of Orders')
plt.xlabel('Number of Orders')
plt.ylabel('Country')
plt.show()

# 3. Average Order Value by Country (Top 10)
revenue_by_country = df_clean.groupby('Country')['TotalPrice'].sum()
orders_count_by_country = df_clean.groupby('Country')['InvoiceNo'].nunique()
avg_order_value = (revenue_by_country / orders_count_by_country).sort_values(ascending=False).head(10)

plt.figure(figsize=(10,5))
sns.barplot(x=avg_order_value.values, y=avg_order_value.index, palette='Spectral')
plt.title('Top 10 Countries by Average Order Value')
plt.xlabel('Average Order Value ()')
plt.ylabel('Country')
plt.show()



```
